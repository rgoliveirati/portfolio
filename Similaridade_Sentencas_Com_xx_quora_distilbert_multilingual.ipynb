```python
#import necessary libraries
import numpy as np
import pandas as pd
```

```python
#load dataset
dataset_types = [ "assin2-train-duas-sentencas.xlsx", "assin2-dev-duas-sentencas.xlsx", "assin2-test-duas-sentencas.xlsx",] # 3 datasets
col_names = ["sentence_1", "sentence_2"] #columns names

#collect as a list of pandas dataframes
df_list = []
for dataset in dataset_types:
  df = pd.read_excel("/content/"+ dataset,
                 header= None,
                 names= col_names
                )
  

  df_list.append(df)

#separate dataset (train, dev and test)
df_train = df_list[0]
df_dev = df_list[1]
df_test = df_list[2]

#show random 5 rows in each dataset (train, dev and test)
print("Show train_data shape: {}".format(df_train.shape))
print("Show dev_data shape: {}".format(df_dev.shape))
print("Show test_data shape: {}".format(df_test.shape))
print("Show Random 5 rows:")
df_train.sample(5)
```

**SIMILARIDADE ENTRE DUAS SENTENÇAS**

```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
title=df_train['sentence_1'].tolist()
claim=df_train['sentence_2'].tolist()
title=model.encode(title)
claim=model.encode(claim)
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(title[0].reshape(1,-1),claim[0].reshape(1,-1))
```

**SIMILARIDADE ENTRE DUAS COLUNAS**

```python
#!pip install spacy_sentence_bert
```

```python
import spacy_sentence_bert
import pandas as pd
nlp = spacy_sentence_bert.load_model('xx_quora_distilbert_multilingual')
df_train_similaridade = df_train
similarityValue = []

for i in range(df_train_similaridade.count()[0]):
    sentence_1 = nlp(df_train_similaridade.iloc[i][0])
    sentence_2 = nlp(df_train_similaridade.iloc[i][1])
    similarityValue.append(sentence_1.similarity(sentence_2))
    #print(sentence_1, '|', sentence_2, '|', sentence_1.similarity(sentence_2))

df_train_similaridade['similaridade'] = similarityValue
```
